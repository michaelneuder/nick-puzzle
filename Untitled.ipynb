```python
import itertools
import pandas as pd
import numpy as np
import copy
```

```python
words = pd.read_csv('words.txt', names=['words'])
words = list(words.words)
len(words)
```

```python
len3Perms = list(itertools.permutations(np.arange(3)))
```

```python
def getWordCombos32k(knowns):
    total = []
    for i in range(2048):
        temp = copy.deepcopy(knowns)
        temp.append(words[i])
        total.append(temp)
    return total

def getWordCombos31k(known):
    total = []
    for i in range(2048):
        for j in range(2048):
            temp = copy.deepcopy(known)
            temp.append(words[i])
            temp.append(words[j])
            total.append(temp)
    return total
```

```python
%%time
res1kgroup1 = getWordCombos32k(["police", "picture"])
res1kgroup2 = getWordCombos32k(["drum", "marriage"])
res1kgroup3 = getWordCombos32k(["face", "light"])
res1kgroup4 = getWordCombos32k(["window", "nature"])
```

```python
len(res1kgroup1)
```

```python
def createMnemonic(group1, group2, group3, group4):
    total = []
    for i in range(6):
        for j in range(6):
            for k in range(6):
                for l in range(6):
                    currG1 = [x for _, x in sorted(zip(len3Perms[i], group1))]
                    currG2 = [x for _, x in sorted(zip(len3Perms[j], group2))]
                    currG3 = [x for _, x in sorted(zip(len3Perms[k], group3))]
                    currG4 = [x for _, x in sorted(zip(len3Perms[l], group4))]
                    temp = []
                    temp.extend(currG1)
                    temp.extend(currG2)
                    temp.extend(currG3)
                    temp.extend(currG4)
                    total.append(temp)
    return total
```

```python
%%time
mnemonics = createMnemonic(["a", "b", "c"], ["a", "b", "c"], ["a", "b", "c"], ["a", "b", "c"])
```
